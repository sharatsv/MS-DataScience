```python
# HW-1
import nltk

nltk.corpus.gutenberg.fileids()

# Get Shakespeare books in the Gutenberh corpus
shakespeare_books = [book for book in nltk.corpus.gutenberg.fileids( ) \
                     if 'shakespeare' in book]

print(shakespeare_books)

# Book-1: Caesar (Genre: Tragedy)
caesar = nltk.corpus.gutenberg.raw(shakespeare_books[0])

# Book-2: Hamlet (Genre: Tragedy/Comedy)
hamlet = nltk.corpus.gutenberg.raw(shakespeare_books[1])
```

```python
# Exercise(2) 
# list the top 50 bigrams by their Mutual Information scores (using min frequency 5)

# Examine the corpus - tokenize, compare the lenght of tokens

# First we will tokenize both the corpus 
caesar_words = nltk.word_tokenize(caesar)
hamlet_words = nltk.word_tokenize(hamlet)

# Compare len of both the corpus
caesar_len = len(caesar_words)
hamlet_len = len(hamlet_words)

# Notice the corpus of words
print('Caeser: %d vs Hamlet: %d' %(caesar_len, hamlet_len))

# Convert corpus to lower-case and remove redundant appearances
caesar_words_uniq = sorted(set([word.lower() for word in caesar_words]))
hamlet_words_uniq = sorted(set([word.lower() for word in hamlet_words]))

print('Caeser(unique): %d vs Hamlet(unique): %d' %(len(caesar_words_uniq), len(hamlet_words_uniq)))
```

```python
# list the top 50 words by frequency (normalized by the length of the document)
from nltk import FreqDist

def freq_dist(words):
    fdist = FreqDist(words)
    fdistkeys = list(fdist.keys())
    print(fdistkeys[:50])
    return fdist
    
# print frequency distribution for each of the corpus
print('Frequency distribution of words from Caesar corpus (raw):')
caesar_fdist = freq_dist(caesar_words)

print('\nFrequency distribution of words from Hamlet corpus (raw):')
hamlet_fdist = freq_dist(hamlet_words)

# First let's see the raw top-50 words
print('\nMost common - top(50) from Caeser corpus:')
for words in caesar_fdist.most_common(50):
    print(words)

print('\nMost common - top(50) from Hamlet corpus:')
for words in hamlet_fdist.most_common(50):
    print(words)

# Convert the words to lower-case
caesar_words = [word.lower() for word in caesar_words]
hamlet_words = [word.lower() for word in hamlet_words]

# As seen in the output, the top word counts also comprise grammar/punctuation marks
# Prior to analyzing the top-50 and normalizing based on overall word counts we will remove these
import re

def is_alpha(word):
    pattern = re.compile('^[a-z]+$')
    if pattern.match(word):
        return True
    else:
        return False
        
def is_alpha_num(word):
    '''
    To fetch numbers/alpha numeric strings
    '''
    pattern = re.compile('^[0-9]+$ | ^[a-z0-9]+$')
    if pattern.match(word):
        return True
    else:
        return False

# Caeser words    
caesar_words_filter = [word for word in caesar_words if is_alpha(word)]
caesar_words_filter.extend([word for word in caesar_words if is_alpha_num(word)])

# Hamlet words
hamlet_words_filter = [word for word in hamlet_words if is_alpha(word)]
hamlet_words_filter.extend([word for word in hamlet_words if is_alpha_num(word)])
```

```python
print('Frequency distribution of words from Caeser corpus (filtered):')
caesar_fdist = freq_dist(caesar_words_filter)

print('\nFrequency distribution of words from Hamlet corpus (filtered):')
hamlet_fdist = freq_dist(hamlet_words_filter)

# First let's see the raw top-50 words
print('\nMost common - top(50) from Caeser corpus:')
for words in caesar_fdist.most_common(50):
    print(words)

print('\nMost common - top(50) from Hamlet corpus:')
for words in hamlet_fdist.most_common(50):
    print(words)
```

```python
def normalize_freq(fdist, count=50):
    total = sum([word[1] for word in fdist.most_common()])
    freq_dict = {word[0]:word[1]/total for word in fdist.most_common()}
    # print the top-N
    for key in list(freq_dict.keys())[:count]:
        print('%s: %s' %(key, freq_dict[key]))

print('\nNormalized by word-frequency - top(50) from Caesar corpus')
normalize_freq(caesar_fdist)

print('\nNormalized by word-frequency - top(50) from Hamlet corpus')
normalize_freq(hamlet_fdist)
```

```python
# Let's import the stopwords and get a true picture of the word-frequencies
nltkstopwords = nltk.corpus.stopwords.words('english')
morestopwords = ['could','would','might','must','need','sha','wo','y',"'s","'d","'ll","'t","'m","'re","'ve", "n't"]

stopwords = nltkstopwords + morestopwords

stopped_caesar_words = [word for word in caesar_words_filter if not word in stopwords]
stopped_hamlet_words = [word for word in hamlet_words_filter if not word in stopwords]

# Print the frequencies again
caesar_fdist = freq_dist(stopped_caesar_words)
hamlet_fdist = freq_dist(stopped_hamlet_words)

print('\nNormalized by word-frequency - top(50) from Caesar corpus')
normalize_freq(caesar_fdist)

print('\nNormalized by word-frequency - top(50) from Hamlet corpus')
normalize_freq(hamlet_fdist)
```

```python
# Exercise 2.2 - List the top 50 bigrams by frequencies
caesar_bigrams = list(nltk.bigrams(caesar_words))
print('Checking for the bi-gram functionality on the Caeser corpus:')
print(caesar_words[:21])
print(caesar_bigrams[:20])

# setup for bigrams and bigram measures
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()

# create the bigram finder and score the bigrams by frequency
def bigram_scores(words, count=50):
    finder = BigramCollocationFinder.from_words(words)
    scored = finder.score_ngrams(bigram_measures.raw_freq)

    # scored is a list of bigram pairs with their score
    for bigram in scored[:count]:
        print(bigram[0], bigram[1])
        
# Let's use the corpus that filtered non-alphabets & removed the stop-words
print('\nBigram measures for top-50 words - Caesar corpus')
bigram_scores(stopped_caesar_words)
print('\nBigram measures for top-50 words - Hamlet corpus')
bigram_scores(stopped_hamlet_words)
```

```python
# Exercise-2.3: List the top 50 bigrams by their Mutual Information scores (using min frequency 5)
def bigram_pmi(words, count=50):
    finder = BigramCollocationFinder.from_words(words)
    
    # For a good correlation apply freq. filter = 5
    finder.apply_freq_filter(5)

    scored = finder.score_ngrams(bigram_measures.pmi)
    scored_dict = {}
    for bscore in scored[:count]:
        scored_dict[bscore[0]] = bscore[1]
        print(bscore)
    return scored_dict

print('PMI for Caesar corpus:')
dict1 = bigram_pmi(stopped_caesar_words, len(stopped_caesar_words))

print('\nPMI for Hamlet corpus:')
dict2 = bigram_pmi(stopped_hamlet_words, len(stopped_hamlet_words))
```
