```python
import matplotlib.pyplot as plt
import numpy as np

# Check for missing by each column
# Let's get data for Arkansas Metro - Hot Springs, Little Rock, Fayetteville, Searcy
# and plot a time-series graph


# First, remove all the rows with NaN/empty-values in Metro/StateName columns
df_eda = df[df['Metro'].isnull() == False]
df_eda = df_eda[df_eda['StateName'].isnull() == False]

# Get state data from Arkansas
df_eda_ar = df_eda.loc[df_eda['State'] == 'AR']


# Filter for specific metros
metros = ['Hot Springs', 'Little Rock', 'Fayetteville', 'Searcy']
                      
ar_hot_springs = df_eda_ar.loc[df_eda_ar['Metro'].str.contains(metros[0])]
ar_little_rock = df_eda_ar.loc[df_eda_ar['Metro'].str.contains(metros[1])]
ar_fayetteville = df_eda_ar.loc[df_eda_ar['Metro'].str.contains(metros[2])]
ar_searcy = df_eda_ar.loc[df_eda_ar['Metro'].str.contains(metros[3])]


# timeseries data
ts_start = df.columns.get_loc('1996-01-31') + 1
ts_end = df.columns.get_loc('2020-03-31')
date_columns = df_eda_ar.iloc[:, ts_start:ts_end].columns.tolist()


# timeseries plot
fig, ax = plt.subplots()
ax.plot(ar_hot_springs[date_columns].mean(), linestyle='solid')
ax.plot(ar_little_rock[date_columns].mean(), linestyle='solid')
ax.plot(ar_fayetteville[date_columns].mean(), linestyle='solid')
ax.plot(ar_searcy[date_columns].mean(), linestyle='solid')

# decrease ticks
xmin, xmax = ax.get_xlim()
ax.set_xticks(np.round(np.linspace(xmin, xmax, 23), 2))

# rotate ticks + show legend
plt.xticks(rotation=90)
plt.gca().legend(('hot_springs', 'little_rock', 'fayetteville', 'searcy'))

# show overall plot
plt.show()

# First analysis - Let's get the min-max and see the % growth over the years
def get_max_percent_growth(df):
    start = df[date_columns]['1996-02-29'].mean()
    finish = df[date_columns]['2020-02-29'].mean()
    growth_percent = (finish - start)/start * 100
    return growth_percent

print('Overall growth % from 1996-2020')
print('-------------------------------')
print('(1) Hot Springs (Blue):', get_max_percent_growth(ar_hot_springs))
print('(2) Little Rock (Orange):', get_max_percent_growth(ar_little_rock))
print('(3) Feyetteville (Green):', get_max_percent_growth(ar_fayetteville))
print('(4) Searcy: (Red): ', get_max_percent_growth(ar_searcy))

# Based on this data Fayetteville yielded max growth in the time-period 1996-2020. However, variation
# should also be considered - see the drop from peak in 2006 till about Sept. 2011. 
#  
```

```python
# Modeling using the ARIMA model
#   - Use 1996-01-31 to 2019-02-28 for train data
#.  - Use 2019-03-31 to 2020-02-29 for test data

# Train-set
train_start = df.columns.get_loc('1996-01-31')
train_end = df.columns.get_loc('2019-02-28')

# Test-set
test_end = df.columns.get_loc('2020-02-29')

# Fetch the corresponding date columns
train_columns = df.iloc[:, train_start:train_end].columns.tolist()
test_columns = df.iloc[:, (train_end + 1):test_end].columns.tolist()

# Create a new df for each
df_train = df[train_columns]
df_test = df[test_columns]

df_train.head()
```

```python
# Let's clean-up all the NaN values in both data-frames
df_train = df_train.dropna()
df_test = df_test.dropna()

df_train.head()
```

```python
from statsmodels.tsa.arima_model import ARIMA

# Transpose the data
df_train = df_train.median().transpose()
df_test = df_test.median().transpose()

# Fit the model
model = ARIMA(df_train, order=(5,1,0))
model_fit = model.fit()
print(model_fit.summary())

# decrease ticks
xmin, xmax = ax.get_xlim()
ax.set_xticks(np.round(np.linspace(xmin, xmax, 23), 2))

# rotate ticks + show legend
plt.xticks(rotation=90)

# Line plot of residuals
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()

# Density plot of residuals
residuals.plot(kind='kde')
plt.show()

# Summary stats of residuals
print(residuals.describe())
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Rolling Forecast ARIMA Model 
#  - use the forecast() function, which performs a one-step forecast using the model.
#. - A rolling forecast is required given the dependence on observations in prior time steps for 
# differencing and the AR model. 
#. - A crude way to perform this rolling forecast is to re-create the ARIMA model after each new 
# observation is received.
#  - We manually keep track of all observations in a list called history that is seeded with the 
# training data and to which new observations are appended in each iteration.
#

history = [i for i in df_train]
predictions = []

# Walk forward validation
for t in range(df_test.shape[0]):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(trend='nc', disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = df_test[t]
    history.append(float(obs))
    print('predicted=%f, expected=%f' % (yhat, obs))

# Evaluate forecasts
rmse = sqrt(mean_squared_error(df_test, predictions))
print('Test RMSE: %.3f' % rmse)

# decrease ticks
xmin, xmax = ax.get_xlim()
ax.set_xticks(np.round(np.linspace(xmin, xmax, 23), 2))

# rotate ticks + show legend
plt.xticks(rotation=90)

# Plot forecasts Vs actual outcomes
plt.plot(df_test)
plt.plot(predictions, color='red')

# The plot shows the predictions in red and the actual values in blue. The two lines eventually converge.
```

```python
import numpy as np

# Let us now move onto grouping the data by Zip-code/RegionName
df_zipcode = df.groupby('RegionName').agg(np.median).dropna().T

# Drop columns like RegionID, SizeRank
df_zipcode_clean = df_zipcode.drop([
    'RegionID',
    'SizeRank'
], axis=0)

# Plot zip-code data
df_zipcode_clean.plot(legend=None)

# decrease ticks
xmin, xmax = ax.get_xlim()
ax.set_xticks(np.round(np.linspace(xmin, xmax, 23), 2))

# rotate ticks + show legend
plt.xticks(rotation=90)

plt.show()

df_zipcode_clean.head()
```

```python
# Let's transpose the df_zipcode - pull the train_columns - transpose it back
predictions_by_zipcode = {}
for column in df_zipcode_clean.T[train_columns].T:
    history = [i for i in df_zipcode_clean[column]]
    predictions = []
    # Walk forward validation
    for t in range(df_zipcode_clean.T[test_columns].T.shape[0]):
        model = ARIMA(history, order=(5,1,0))
        model_fit = model.fit(trend='nc', disp=0)
        output = model_fit.forecast()
        yhat = output[0][0]
        predictions.append(yhat)
        obs = df_zipcode_clean.T[test_columns].T[column][t]
        history.append(float(obs))
        # Print for first & last iteration only
        if (column == 1001 or column == 99508) and t == 0:
            print('predicted=%f, expected=%f' % (yhat, obs))
    predictions_by_zipcode[column] = predictions
```
